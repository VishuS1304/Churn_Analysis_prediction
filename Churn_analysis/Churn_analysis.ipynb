**Data Cleaning:-**

**Make copy of original dataset for manupulation and processing:-**

```python
churn_data = churn_dataset.copy()
```

**Total Charges should be numeric amount. Convert it to numerical data type:-**

```python
churn_data.TotalCharges = pd.to_numeric(churn_data.TotalCharges, errors= 'coerce')
churn_data.isnull().sum()
```

**As we can see there are 11 missing values. So by checking these records through location:-**

```python
churn_data.loc[churn_data['TotalCharges'].isnull() == True]
```

**Missing Value Treatement:-**

```python
# Removing missing value
churn_data.dropna(how = 'any', inplace = True)
# Get the max tenure 
print(churn_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0}-{1}".format(i,i+11) for i in range(1,72,12)] 
churn_data['tenure_group'] = pd.cut(churn_data.tenure, range(1,80,12), right = False, labels = labels )
```

```python
churn_data['tenure_group'].value_counts()
```

**Remove columns that are not required for processing:-**

```python
# Drop column customerID and tenure
churn_data.drop(columns =['customerID','tenure'], axis=1, inplace = True)
churn_data.head()
```

**Data Exploration:-**

Plot distribution of Individual predictors by Churn

```python
for i , predictor in enumerate(churn_data.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
  plt.figure(i)
  sns.countplot(data=churn_data , x=predictor , hue='Churn')
```

Convert the target varible 'Churn' in binary numeric variable that is Yes=1; No=0;

```python
churn_data['Churn']= np.where(churn_data.Churn == 'Yes',1,0)
```

```python
churn_data.head()
```

Convert all the categorical variables into dummy variables(or in numeric):-

```python
churn_data_dummies= pd.get_dummies(churn_data)
churn_data_dummies.head()
```

Churn by Monthly Charges and Total Charges:-

```python
Mth = sns.kdeplot(churn_data_dummies.MonthlyCharges[(churn_data_dummies["Churn"] == 0) ],color="Red", fill = True)
Mth = sns.kdeplot(churn_data_dummies.MonthlyCharges[(churn_data_dummies["Churn"] == 1) ],ax =Mth, color="Blue", fill= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

**Insight:**
Churn is high when Monthly Charges are high

```python
Tot = sns.kdeplot(churn_data_dummies.TotalCharges[(churn_data_dummies["Churn"] == 0) ],color="Red", fill = True)
Tot = sns.kdeplot(churn_data_dummies.TotalCharges[(churn_data_dummies["Churn"] == 1) ],ax =Tot, color="Blue", fill= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

**Surprising Insight** as higher Churn at lower Total Charges

**Higher Monthly Charge, Lower tenure and Lower Total Charge are linkd to High Churn**

**Bivariate Analysis:-**

```python
new_df1_target0=churn_data.loc[churn_data["Churn"]==0]
new_df1_target1=churn_data.loc[churn_data["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='Contract',title='Distribution of Contract for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='TechSupport',title='Distribution of TechSupport for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='SeniorCitizen',title='Distribution of SeniorCitizen for Churned Customers',hue='gender')
```
